10. What is the difference between map(), reduce(), and filter() functions in Python?
Function	Purpose	Returns
map()	Applies a function to each element	Map object
filter()	Filters elements by a condition	Filter object
reduce()	Applies rolling computation	Single result
Example:
from functools import reduce

print(list(map(lambda x: x*2, [1, 2, 3])))      # [2, 4, 6]
print(list(filter(lambda x: x > 1, [1, 2, 3]))) # [2, 3]
print(reduce(lambda x, y: x + y, [1, 2, 3]))    # 6

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(x):
  for i in range(2,x):
    if x%i == 0:
      return 'Not prime'
    else:
      return 'prime'
print(prime(3))
print(prime(67))
print(prime(112))
print(prime(273))
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
n= int(input('give the number: '))
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fibonacci_generator(n)
for i in fibonacci_generator(n):
  print(i)





```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power(x):
  for i in range(x+1):
    yield 2**i
y = iter(power(5))
for i in y:   ## instead of writing each time next wrote in for loop
  print(i)


```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def temp(c):
  F = (c * 9/5) + 32
  return F
temp(36)

l = [23,56,123,45,39]

fahrenheit_temps=list(map(temp,l))
print(fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def vowel(x):
    for i in x:
      if i not in ('a','e','i','o','u','A','E','I','O','U'):
        return True
x = 'SAdhana'

withou_vowel = list(filter(vowel,x))
print(withou_vowel)

x = 'SAdhana'
only_consonants= list(filter(lambda y: y not in ('a','e','i','o','u','A','E','I','O','U'), x))
print(''.join(only_consonants))
```
