Data Toolkit Assignment Theory Questions

1. What is NumPy, and why is it widely used in Python?
NumPy (Numerical Python) is a Python library for fast numerical computations, especially with multidimensional arrays. It’s widely used because it offers high performance (written in C), supports mathematical and statistical functions, and enables vectorized operations for efficiency.

2. How does broadcasting work in NumPy?
Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array along dimensions without making copies.

3. What is a Pandas DataFrame?
A DataFrame is a two-dimensional, tabular data structure in Pandas with labeled rows and columns, similar to a spreadsheet or SQL table.

4. Explain the use of the groupby() method in Pandas.
groupby() is used to split data into groups based on column values, apply functions (like sum, mean, count), and combine the results, enabling aggregation and summarization.

5. Why is Seaborn preferred for statistical visualizations?
Seaborn provides a high-level API for creating aesthetically pleasing statistical plots with minimal code, built on top of Matplotlib, and integrates well with Pandas.

6. What are the differences between NumPy arrays and Python lists?
•	NumPy arrays are fixed-type, while Python lists can store mixed types.
•	Arrays are faster and use less memory.
•	Arrays support vectorized operations; lists require loops.

7. What is a heatmap, and when should it be used?
A heatmap is a graphical representation of data where values are represented by colors. It’s used to show correlation matrices, intensity patterns, or geographical data density.

8. What does the term “vectorized operation” mean in NumPy?
It refers to performing operations on entire arrays at once without explicit Python loops, resulting in faster execution.

9. How does Matplotlib differ from Plotly?
•	Matplotlib: static, highly customizable, traditional plotting library.
•	Plotly: interactive, supports zooming/hovering, works well for dashboards and web apps.

10. What is the significance of hierarchical indexing in Pandas?
Hierarchical (multi-level) indexing allows multiple index levels in rows or columns, enabling more complex data representations and easier slicing.

11. What is the role of Seaborn’s pairplot() function?
pairplot() automatically plots pairwise relationships between numerical variables in a dataset, along with histograms or KDEs on the diagonal.

12. What is the purpose of the describe() function in Pandas?
It generates summary statistics (count, mean, std, min, quartiles, max) for numerical columns in a DataFrame.

13. Why is handling missing data important in Pandas?
Missing data can lead to inaccurate analysis, errors in calculations, and biased results, so it’s essential to detect, fill, or drop them.

14. What are the benefits of using Plotly for data visualization?
Plotly creates interactive, high-quality plots with minimal code, supports multiple chart types, works well in browsers, and integrates with Dash for dashboards.

15. How does NumPy handle multidimensional arrays?
It stores them in a contiguous memory block and uses shape and strides to interpret dimensions, enabling fast access and operations.

16. What is the role of Bokeh in data visualization?
Bokeh is a Python library for creating interactive, web-ready visualizations that can be embedded in HTML, supporting streaming and real-time data.

17. Explain the difference between apply() and map() in Pandas.
•	map() works on Series, element-wise.
•	apply() works on Series or DataFrames, allowing row/column-wise or element-wise custom functions.

18. What are some advanced features of NumPy?
Advanced indexing, broadcasting, memory mapping of large files, masked arrays, FFT, linear algebra operations, random number generation, and structured arrays.

19. How does Pandas simplify time series analysis?
It offers date parsing, frequency conversion, resampling, rolling windows, shifting, and easy plotting for time-indexed data.

20. What is the role of a pivot table in Pandas?
A pivot table summarizes data by grouping and aggregating, rearranging it into a matrix-like format for easier analysis.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy arrays are stored in contiguous memory and use optimized C-based operations, while Python lists store references, requiring more overhead.

22. What are some common use cases for Seaborn?
•	Visualizing correlations with heatmaps
•	Comparing distributions (histograms, KDEs)
•	Analyzing relationships (scatter, pairplot)
•	Showing categorical comparisons (bar, box, violin plots)

```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Math': np.random.randint(50,100,10),
        'English' : np.random.randint(50,100,10),
        'science': np.random.randint(50,100,10)}
df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)

# to get heatmap
sns.heatmap(corr_matrix, annot = True, cmap='coolwarm',fmt = '.2f')
```
