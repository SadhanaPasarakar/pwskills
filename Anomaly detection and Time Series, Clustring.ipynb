Question 1: What is Dimensionality Reduction? Why is it important in machine learning?
Answer:
Dimensionality Reduction is the process of reducing the number of input features while preserving as much relevant information as possible.
Importance:
•	Reduces computational cost
•	Mitigates the curse of dimensionality
•	Removes noise and redundant features
•	Helps visualization in 2D or 3D
•	Improves model performance and generalization

Question 2: Name and briefly describe three common dimensionality reduction techniques.
Answer:
1.	PCA (Principal Component Analysis):
Transforms data into new axes (principal components) that capture maximum variance.
2.	t-SNE (t-Distributed Stochastic Neighbor Embedding):
Non-linear method used mainly for visualization by preserving local structure.
3.	LDA (Linear Discriminant Analysis):
Supervised technique that finds feature combinations maximizing class separation.

Question 4: Explain the concept of anomaly detection and its significance.
Answer:
Anomaly detection identifies rare events or patterns that deviate significantly from normal behavior.
Significance:
•	Detect fraud, system failures, cyber-attacks
•	Maintain system reliability
•	Identify unusual trends in business or healthcare
________________________________________
Question 5: List and briefly describe three types of anomaly detection techniques.
Answer:
1.	Statistical Methods:
Use probability distributions or thresholds to flag unusual values.
2.	Distance-based Methods:
Detect anomalies by measuring distance from normal data (e.g., k-NN outlier detection).
3.	Machine Learning Methods:
Use models such as Isolation Forest, Autoencoders, or One-Class SVM to learn normal patterns.

Question 6: What is time series analysis? Mention two key components of time series data.
Answer:
Time series analysis involves studying data points collected over time to identify patterns and make predictions.
Key components:
•	Trend: Long-term upward or downward movement.
•	Seasonality: Regular repeating patterns at fixed intervals.
________________________________________
Question 7: Describe the difference between seasonality and cyclic behavior in time series.
Answer:
•	Seasonality:
Repeats at regular, known intervals (daily, weekly, yearly).
•	Cyclic behavior:
Fluctuations that occur over irregular, longer periods (e.g., economic cycles) and do not have fixed lengths.

Question 10: How can time series analysis be used for anomaly detection?
Answer:
Time series analysis detects anomalies by identifying unexpected deviations in temporal patterns.
Methods include:
•	Comparing actual values vs. predicted values (forecasting models like ARIMA, LSTM)
•	Detecting sudden spikes, drops, or irregular seasonal patterns
•	Using statistical thresholds on residual errors
This helps detect fraud, sensor failures, unusual traffic spikes, etc.

```python
# Question 8: Write Python code to perform K-means clustering on a sample dataset.
# Answer:


from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X')
plt.title("K-Means Clustering Output")
plt.show()
```

```python
# Question 9: What is inheritance in OOP? Provide a simple example in Python.
# Answer:
# Inheritance allows one class (child) to acquire the attributes and methods of another class (parent).
# It promotes code reuse and hierarchical structure.
# Example:
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        return "Dog barks"

d = Dog()
print(d.speak())  # Output: Dog barks
```
